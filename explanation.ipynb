## Session Timeline Example

**1st Question:**
```
What gets sent to LLM:
[
    SystemMessage("You are a helpful assistant..."),
    HumanMessage("Hi, my name is Sarah")  ← Only this question
]

LLM Response: "Hello Sarah! Nice to meet you."

History now contains:
- HumanMessage("Hi, my name is Sarah")  
- AIMessage("Hello Sarah! Nice to meet you.")
```

**2nd Question:**
```
What gets sent to LLM:
[
    SystemMessage("You are a helpful assistant..."),
    HumanMessage("Hi, my name is Sarah"),           ← Previous question
    AIMessage("Hello Sarah! Nice to meet you."),    ← Previous response
    HumanMessage("I like chocolate")                ← New question
]

LLM Response: "That's great! Chocolate is delicious."

History now contains all 4 messages above.
```

**3rd Question:**
```
What gets sent to LLM:
[
    SystemMessage("You are a helpful assistant..."),
    HumanMessage("Hi, my name is Sarah"),           ← Question 1
    AIMessage("Hello Sarah! Nice to meet you."),    ← Response 1
    HumanMessage("I like chocolate"),               ← Question 2
    AIMessage("That's great! Chocolate is delicious."), ← Response 2
    HumanMessage("What's my name?")                 ← New question
]

LLM Response: "Your name is Sarah."
```

## Key Points

**Cumulative History**: Each interaction adds to the growing conversation history. The LLM always receives the complete conversation from the beginning.

**No Selective Memory**: The LLM doesn't just get "recent" messages - it gets everything from that session.

**Fresh Processing**: The LLM processes the entire conversation as if seeing it for the first time each round. It doesn't retain anything between calls.

**Token Cost**: This means longer conversations use more tokens since you're sending the full history each time.

**Context Preservation**: This approach ensures the LLM has full context for every response, which is why it can reference things said earlier in the conversation.

This is exactly how systems like ChatGPT work - they maintain the illusion of memory by always sending the complete conversation history with each new message.
